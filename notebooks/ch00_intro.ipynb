## Getting Your Environment Ready

### How Colab environments work — and why this matters

Before running the setup cells, it is worth understanding how Google Colab works under the hood, because it directly shapes how this book is organised.

When you open a notebook in Colab, Google assigns you a **temporary virtual machine** — a fresh Linux computer running in the cloud. It comes with Python pre-installed, but most data science and deep learning packages are not there yet. When you install a package (e.g. `pip install torch`), it is installed on *that* virtual machine, for *that* session only.

**When the session ends — or when you close and reopen any notebook — the virtual machine is discarded.** All installed packages disappear. The next time you open a notebook, you start again with a clean machine.

This has one important consequence for how you use this book:

> **Every chapter notebook installs its own dependencies in its very first cell.** You do not need to return to this Chapter 0 notebook before opening a chapter. Each notebook is fully self-contained — open it, run the first cell, and everything it needs will be installed for that session automatically.

The setup cells in *this* chapter serve a different purpose: **they are a learning exercise.** Running them here teaches you exactly what a setup cell does, what correct output looks like, and how to diagnose a problem — before you encounter one in a chapter where you also have new concepts to focus on.

If you are running on a **local machine** (not Colab), packages persist between sessions once installed. The same cells work in both environments.

### Step 1 — Install packages

````python
# ─────────────────────────────────────────────────────────────────────────────
# STEP 1 — Install packages (Learning Exercise)
#
# This cell installs all packages used across the full book so you can
# see the pattern in one place and verify everything works.
#
# In each chapter notebook, the first cell installs only the packages
# that chapter actually needs. The pattern is identical to what you
# see here — just a shorter list.
#
# On Colab:         packages install in ~2–5 minutes. They are lost
#                   when the session ends — each chapter reinstalls its own.
# On local Jupyter: packages persist after the first install.
# ─────────────────────────────────────────────────────────────────────────────

import subprocess, sys

packages = [
    "torch",           # Deep learning framework (Chapters 2–6)
    "torchvision",     # Image datasets and transforms (Chapters 4–5)
    "torchtext",       # Text utilities (Chapter 6)
    "numpy",           # Numerical computing
    "pandas",          # Tabular data handling
    "matplotlib",      # Plotting and visualisation
    "seaborn",         # Statistical plots
    "scikit-learn",    # Preprocessing, metrics, classical ML baselines
    "transformers",    # Hugging Face — tokenizers and models (Chapters 6–7)
    "datasets",        # Hugging Face — dataset loading (Chapters 6–7)
    "gradio",          # Simple web interfaces for demos (Chapter 7)
    "kaggle",          # Kaggle API for dataset downloads (Chapters 4–6)
    "ipywidgets",      # Interactive widgets in notebooks
    "tqdm",            # Progress bars during training
]

print("Installing packages...")
print("Each line shows the package name and whether it installed successfully.\n")

for pkg in packages:
    result = subprocess.run(
        [sys.executable, "-m", "pip", "install", "--quiet", pkg],
        capture_output=True, text=True
    )
    status = "OK" if result.returncode == 0 else "FAILED"
    print(f"  {pkg:<20} {status}")

print("\nDone. If any package shows FAILED, re-run this cell once.")
print("Proceed to Step 2.")
````

### Step 2 — Verify your environment

This cell imports each library and prints its installed version. It also reports the available compute device. **Seeing `CPU only` is perfectly normal** — a GPU is not required for this book.

````python
# ─────────────────────────────────────────────────────────────────────────────
# STEP 2 — Verify environment
# Imports each library and prints its installed version.
# Reports CPU, CUDA GPU, or Apple MPS — whichever is available.
# ─────────────────────────────────────────────────────────────────────────────

import sys
import importlib

print("=" * 55)
print("  Environment Verification")
print("=" * 55)

# Python version
py_version = sys.version.split(" ")[0]
py_ok      = int(py_version.split(".")[1]) >= 9
py_status  = "OK" if py_ok else "WARNING — Python 3.9+ recommended"
print(f"  {'Python':<16} {py_version:<12}  {py_status}")

# Core libraries
libs = [
    "torch",
    "torchvision",
    "numpy",
    "pandas",
    "matplotlib",
    "sklearn",
    "transformers",
    "gradio",
]

for lib_name in libs:
    try:
        mod     = importlib.import_module(lib_name)
        version = getattr(mod, "__version__", "unknown")
        print(f"  {lib_name:<16} {version:<12}  OK")
    except ImportError:
        print(f"  {lib_name:<16} {'NOT FOUND':<12}  Re-run Step 1")

# Compute device
print()
print("-" * 55)
try:
    import torch
    if torch.cuda.is_available():
        gpu_name = torch.cuda.get_device_name(0)
        print(f"  Compute          GPU — {gpu_name}")
        print(f"  CUDA version     {torch.version.cuda}")
    elif hasattr(torch.backends, "mps") and torch.backends.mps.is_available():
        print("  Compute          Apple MPS (Metal GPU)")
    else:
        print("  Compute          CPU only")
        print("                   This is fine. All exercises run correctly on CPU.")
        print("                   GPU speeds up training in Ch. 4-6 but is not required.")
except Exception as e:
    print(f"  Compute check failed: {e}")

print("=" * 55)
print("  Verification complete. Proceed to Step 3.")
print("=" * 55)
````

### Step 3 — PyTorch smoke test

This cell creates two small matrices and multiplies them. **Don't worry about understanding what is happening here** — matrix multiplication is covered in detail in Chapter 2. The only goal right now is to confirm that PyTorch is installed and running correctly on your device.

````python
# ─────────────────────────────────────────────────────────────────────────────
# STEP 3 — PyTorch smoke test
# Creates two matrices and multiplies them.
# If this runs without error, PyTorch is working correctly.
# Matrix multiplication is explained fully in Chapter 2.
# ─────────────────────────────────────────────────────────────────────────────

import torch

# Detect the best available device
if torch.cuda.is_available():
    device = torch.device("cuda")
elif hasattr(torch.backends, "mps") and torch.backends.mps.is_available():
    device = torch.device("mps")
else:
    device = torch.device("cpu")

print(f"Running on: {device}")
print()

# Two simple 2x2 matrices
A = torch.tensor([[1.0, 2.0],
                   [3.0, 4.0]], device=device)

B = torch.tensor([[5.0, 6.0],
                   [7.0, 8.0]], device=device)

C = torch.matmul(A, B)  # Matrix multiplication — covered in Chapter 2

print("Matrix A:")
print(A)
print()
print("Matrix B:")
print(B)
print()
print("A x B =")
print(C)
print()
print("Smoke test passed. PyTorch is working correctly.")
````

## A Recurring Pattern: The Model Pipeline

One of the most important professional habits this book builds is knowing what to do **after** a model is trained. In the real world, a trained model is only useful if it can be saved, handed off, deployed, and updated with new data. This book treats that as a first-class skill — not an afterthought.

Starting in **Chapter 3**, every model we build is wrapped in a `ModelPipeline` — a Python class that bundles everything needed to reproduce the model's predictions or continue its training:

| What is saved | Why it matters |
|---------------|----------------|
| **Model architecture config** | Enables rebuilding the exact same network on load — no need to redefine it |
| **Model weights** (`state_dict`) | The learned parameters — saved using PyTorch's standard `state_dict` approach |
| **Preprocessor** | The scaler, encoder, or image transforms fitted on training data |
| **Feature names and types** | Used to validate incoming data before prediction or retraining |
| **Training history** | Loss and metric curves — useful for diagnosing and continuing training |

Everything is saved into a single **`.pth` file** using `torch.save()`. Loading it in any session — on any machine — reconstructs a fully working pipeline with one line of code.

> **Why `state_dict` and not `torch.save(model)`?**  
> Saving the full model object with `torch.save(model)` uses Python's pickle and stores a *reference* to the class definition, not the definition itself. When you open that file in a new session, Python cannot find the class and raises an error. Saving `state_dict` (weights only) alongside the architecture config (a plain dictionary) avoids this entirely — it works reliably across sessions, machines, and environments. This is the PyTorch-recommended approach and the only one used in this book.

**Input validation** is built into the pipeline. Before any prediction or retraining call, the pipeline checks that the incoming data matches what the model was originally trained on — correct feature names, correct types, no unexpected nulls, sensible value ranges. Clear error messages tell you exactly what is wrong rather than letting the model silently produce incorrect results.

Chapters 3 through 6 each close with the same pattern applied to a different architecture:

| Chapter | Architecture | Pipeline demonstrates |
|---------|-------------|----------------------|
| 3 | Regression network | Introduction of `ModelPipeline` — save, load, predict, retrain |
| 4 | Feedforward classifier | Pipeline with tabular data and class labels |
| 5 | CNN | Pipeline with image transforms and shape validation |
| 6 | LSTM | Pipeline with sequence data and windowing |

By the end of Chapter 6, saving and deploying a model will feel as natural as training one.
