# 1.1 The Business Case for Deep Learning

## Why now?

Deep learning is not new. The mathematical foundations — artificial neurons, backpropagation — were worked out in the 1980s. For two decades, the ideas existed but the results were disappointing. Then, roughly between 2010 and 2015, something changed. Deep learning began winning competitions, beating human benchmarks, and powering products used by billions of people.

Three forces converged to make this happen:

---

### Force 1: Data

Deep learning is data-hungry. A model with millions of parameters needs millions of examples to learn anything useful. Before the smartphone era, that data simply did not exist in digital form. By 2015, humanity was generating more data every two days than had existed in all of human history up to 2003. Images, text, transactions, sensor readings, clickstreams — the raw material deep learning needed was suddenly everywhere.

### Force 2: Compute

Training a large neural network is a massive arithmetic exercise — billions of multiplications and additions, repeated thousands of times. Graphics Processing Units (GPUs), originally designed for video games, turned out to be extraordinarily well-suited to this task. A GPU can perform thousands of arithmetic operations simultaneously, making training that would have taken months on a CPU feasible in hours.

### Force 3: Frameworks

The third force is often underappreciated: **software**. Libraries like PyTorch and TensorFlow made it possible to build and train complex neural networks in dozens of lines of code rather than thousands. They handle the difficult mathematics automatically — you describe what you want to build, and the framework works out how to train it. This is why you do not need a PhD to work through this book.

---

## Where does deep learning fit?

Deep learning is not a replacement for everything that came before it. It is a powerful addition to the toolkit — one that excels at specific types of problems.

**Table 1.1 — The Analytics Toolkit: Three Complementary Approaches**

| | Classical Statistics | Classical Machine Learning | Deep Learning |
|--|---------------------|--------------------------|---------------|
| **Examples** | Linear regression, hypothesis testing, ARIMA | Decision trees, random forests, XGBoost, SVMs | Neural networks, CNNs, LSTMs, Transformers |
| **Data type** | Structured (tables) | Structured (tables) | **Both structured and unstructured** |
| **Data size** | Small | Small to medium | Medium to large |
| **Best for** | Inference, significance testing, forecasting with equations | Tabular prediction, fast training, strong baselines | Images, text, audio, sequences — and large structured datasets with complex patterns |
| **Interpretability** | High | Medium to high | Generally lower |
| **Hardware** | CPU | CPU | GPU helps significantly |

> **Important:** Deep learning works on both structured (tabular) and unstructured data. In this book, Chapter 4 uses deep learning on tabular customer data (churn prediction), while Chapters 5 and 6 tackle images and sequences. Classical ML typically wins on small structured datasets; deep learning adds most value when data is large, unstructured, or has very complex patterns.

These three approaches are **complementary, not competing**. A business analytics professional uses all three — choosing the right tool for each problem.

# 1.2 What Is an Artificial Neural Network?

## The biological inspiration

The human brain contains roughly 86 billion neurons. Each neuron receives electrical signals from many other neurons through connections called synapses. When the combined incoming signal is strong enough, the neuron fires — sending its own signal forward to the neurons it is connected to. Learning, in a biological brain, happens by strengthening some connections and weakening others.

An **artificial neural network (ANN)** is a very loose computational analogy to this. It is important not to take the analogy too seriously — an artificial neuron is a simple mathematical function, not a biological cell. But the inspiration shapes the vocabulary we use.

---

## The artificial neuron

A single artificial neuron does exactly three things:

1. **Receives inputs** — numbers representing features of the data (e.g. age, income, pixel brightness)
2. **Computes a weighted sum** — multiplies each input by a learned weight and adds a bias
3. **Applies an activation function** — squashes the result into a useful range and introduces non-linearity

Mathematically:

$$\text{output} = f\left(\sum_{i} w_i x_i + b\right)$$

Where $x_i$ are the inputs, $w_i$ are the weights, $b$ is the bias, and $f$ is the activation function.

**Weights** control how much attention the neuron pays to each input.  
**Bias** allows the neuron to shift its output up or down, independent of the inputs.  
**Activation functions** are covered in depth in Chapter 3. For now, think of them as a way to introduce non-linearity — without them, stacking multiple layers would be mathematically equivalent to a single layer.

```python
# ─────────────────────────────────────────────────────────────────────────────
# A single artificial neuron — implemented from scratch
#
# We will manually compute the output of one neuron given:
#   - 3 inputs (e.g. age, income, years as customer)
#   - 3 weights (one per input)
#   - 1 bias
#   - ReLU activation function (returns 0 if negative, otherwise returns the value)
#
# This is exactly what PyTorch does inside every nn.Linear layer.
# ─────────────────────────────────────────────────────────────────────────────

import numpy as np

# --- Inputs: one customer's normalised feature values ---
# (normalisation is covered in Chapter 2; for now these are just numbers)
inputs = np.array([0.6,   # normalised age
                   0.8,   # normalised income
                   0.3])  # normalised tenure (years as customer)

# --- Weights: how much this neuron cares about each input ---
# In a real network these start random and are learned during training
weights = np.array([0.4,   # weight for age
                    0.9,   # weight for income
                   -0.2])  # weight for tenure (negative = higher tenure -> lower output)

# --- Bias: a fixed offset added after the weighted sum ---
bias = 0.1

# --- Step 1: Weighted sum ---
# Each input is multiplied by its weight, then all products are summed, then bias is added
weighted_sum = np.dot(inputs, weights) + bias

print("Step 1: Weighted sum")
for i, (x, w) in enumerate(zip(inputs, weights)):
    print(f"  Input {i+1}: {x} × {w} = {x*w:.4f}")
print(f"  Bias:               + {bias}")
print(f"  Weighted sum total:   {weighted_sum:.4f}")

# --- Step 2: Activation function (ReLU) ---
# ReLU (Rectified Linear Unit): max(0, x)
# If the weighted sum is negative, output is 0. Otherwise, output is the value itself.
def relu(x):
    # max(0, x) — the simplest and most commonly used activation function
    return max(0, x)

neuron_output = relu(weighted_sum)

print(f"\nStep 2: ReLU activation")
print(f"  ReLU({weighted_sum:.4f}) = {neuron_output:.4f}")
print(f"\nNeuron output: {neuron_output:.4f}")
print("\nThis single number is what gets passed to the next layer.")
```

## From one neuron to a network

A single neuron can only represent a simple relationship between inputs and output. The power of neural networks comes from arranging many neurons into **layers** and stacking those layers.

Every neural network has three types of layers:

| Layer | Role |
|-------|------|
| **Input layer** | Receives the raw feature values — one node per feature. No computation happens here. |
| **Hidden layers** | Perform the actual learning. Each neuron combines its inputs, applies weights, and passes results forward. The word "deep" in deep learning refers to having multiple hidden layers. |
| **Output layer** | Produces the final prediction. One node for regression (a single number), one node per class for classification. |

```python
# ─────────────────────────────────────────────────────────────────────────────
# Figure 1.2 — Anatomy of a Neural Network
# Visualises a small network: 3 inputs → 4 hidden → 4 hidden → 1 output
# ─────────────────────────────────────────────────────────────────────────────

fig, ax = plt.subplots(figsize=(13, 7))
ax.set_xlim(-0.5, 10.5)
ax.set_ylim(-0.5, 6.5)
ax.axis('off')
fig.patch.set_facecolor('white')
ax.set_facecolor('white')

# Define layer positions and neuron counts
layers = [
    {'x': 1.5,  'n': 3, 'label': 'Input Layer',    'sublabel': '3 features',      'color': '#aed6f1', 'edge': '#2980b9'},
    {'x': 4.0,  'n': 4, 'label': 'Hidden Layer 1', 'sublabel': '4 neurons',       'color': '#a9dfbf', 'edge': '#27ae60'},
    {'x': 6.5,  'n': 4, 'label': 'Hidden Layer 2', 'sublabel': '4 neurons',       'color': '#a9dfbf', 'edge': '#27ae60'},
    {'x': 9.0,  'n': 1, 'label': 'Output Layer',   'sublabel': '1 prediction',    'color': '#f9e79f', 'edge': '#d4ac0d'},
]

neuron_positions = []

for layer in layers:
    x = layer['x']
    n = layer['n']
    # Centre the neurons vertically
    y_positions = [3.0 + (i - (n-1)/2) * 1.2 for i in range(n)]
    neuron_positions.append(list(zip([x]*n, y_positions)))

    # Draw connections to previous layer
    if len(neuron_positions) > 1:
        for x2, y2 in neuron_positions[-1]:
            for x1, y1 in neuron_positions[-2]:
                ax.plot([x1, x2], [y1, y2], color='#cccccc', lw=0.8, zorder=1)

    # Draw neurons
    for x_n, y_n in zip([x]*n, y_positions):
        circle = plt.Circle((x_n, y_n), 0.38, color=layer['color'],
                             ec=layer['edge'], lw=2, zorder=2)
        ax.add_patch(circle)

    # Layer labels
    ax.text(x, 0.2, layer['label'],    ha='center', fontsize=10, fontweight='bold', color='#1F3864')
    ax.text(x, -0.2, layer['sublabel'], ha='center', fontsize=9,  color='#555555', fontstyle='italic')

# Annotate input nodes
input_labels = ['Age', 'Income', 'Tenure']
for (x, y), label in zip(neuron_positions[0], input_labels):
    ax.text(x - 0.6, y, label, ha='right', va='center', fontsize=9, color='#1a5276')

# Annotate output node
ax.text(9.7, 3.0, 'Churn\nRisk', ha='left', va='center', fontsize=9, color='#7d6608')

# Annotate one neuron with a callout — no colour change, just a label
hx, hy = neuron_positions[1][1]   # second neuron in hidden layer 1
ax.annotate('Each circle = one neuron\n(weighted sum + activation)',
            xy=(hx, hy), xytext=(hx - 0.2, hy + 2.1),
            fontsize=9, color='#1F3864',
            arrowprops=dict(arrowstyle='->', color='#1F3864', lw=1.5),
            ha='center')

ax.set_title('Figure 1.2 — Anatomy of a Neural Network\n'
             'Inputs flow left to right through hidden layers to produce a prediction',
             fontsize=12, fontweight='bold', color='#1F3864', pad=15)
plt.tight_layout()
plt.show()
```

# 1.3 Machine Learning vs. Deep Learning

Deep learning is a **subset** of machine learning — not a replacement for it. Understanding when to use which approach is one of the most practical skills a business analytics professional can develop.

## Three important clarifications

### 1. Feature engineering still exists in deep learning — just less of it

In **classical machine learning**, a human expert manually designs the features fed to the model. For a customer churn model, you might create features like: *number of support calls in the last 30 days*, *change in usage from last month*, *days since last login*. This takes significant domain expertise and time.

Deep learning **reduces** this burden but does not eliminate it. For structured (tabular) data — like the churn prediction model in Chapter 4 — you still select which columns to use, handle missing values, encode categorical variables, and normalise numeric ranges. The difference is that the network then learns complex interactions between those features on its own, without you specifying them manually.

For unstructured data — images, raw text — deep learning almost eliminates feature engineering. A CNN learns its own visual features (edges, textures, shapes) directly from pixels. There is no practical way for a human to design those features manually.

### 2. Deep learning models are deeper — that is where the name comes from

Classical ML models like decision trees, random forests, and SVMs are typically **shallow**: one or a few computational layers. A deep learning model has **many layers** — anywhere from a handful to hundreds. This depth is what allows DL models to learn hierarchical representations: early layers detect simple patterns, later layers combine them into complex ones.

### 3. Where does a simple ANN sit?

A basic feedforward neural network with one or two hidden layers sits at the boundary. Technically, any neural network is deep learning by architecture. Practically, a very shallow network (one hidden layer, a handful of neurons) is closer to classical ML in its capabilities and behaviour — it can only learn simple patterns. When practitioners say "deep" learning, they typically mean networks with enough layers to learn complex, hierarchical representations. In this book, we start shallow and build up.

---

## Comparison table

| | Classical ML | Deep Learning |
|--|-------------|---------------|
| **Feature engineering** | Mostly manual | Reduced — but still needed for structured data |
| **Model depth** | Shallow (1–2 layers) | Deep (many layers) |
| **Data needed** | Hundreds to thousands | Thousands to millions |
| **Training time** | Seconds to minutes | Minutes to hours |
| **Interpretability** | Generally high | Generally lower |
| **Best data type** | Structured (tables) | Both structured and unstructured |
| **Hardware** | Standard CPU | GPU significantly helps |
| **When it wins** | Small-medium tabular data | Images, text, sequences, large/complex datasets |

# 1.4 Forward & Backward Propagation

Training a neural network means finding the right values for all its weights and biases. A network with two hidden layers of 64 neurons each has tens of thousands of parameters. How does it find values that produce good predictions?

The answer is a two-phase cycle that repeats for every batch of training data:

1. **Forward pass** — data flows through the network from input to output, producing a prediction
2. **Backward pass (backpropagation)** — the prediction error flows backward through the network, adjusting every weight slightly to reduce that error

---

## Phase 1: The Forward Pass

The forward pass is straightforward. Each layer takes its inputs, multiplies them by its weights, adds a bias, applies an activation function, and passes the result to the next layer. The final layer produces a prediction.

This is just arithmetic — no learning happens in the forward pass. It is purely calculation.

```python
# ─────────────────────────────────────────────────────────────────────────────
# A manual forward pass through a tiny 2-layer network
#
# Network architecture:
#   Input:    2 features  (e.g. hours studied, previous exam score)
#   Layer 1:  3 neurons   (hidden)
#   Layer 2:  1 neuron    (output — predicted exam score)
#
# We use fixed weights here so the output is predictable.
# In a real network, weights start random.
# ─────────────────────────────────────────────────────────────────────────────

import numpy as np

# Activation function: ReLU
def relu(x):
    return np.maximum(0, x)

# Activation function: Sigmoid (squashes any number to [0, 1] — good for probabilities)
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

# --- Input: one student's data ---
X = np.array([0.7,   # hours studied (normalised)
              0.85]) # previous score (normalised)

# --- Layer 1 weights and bias (2 inputs → 3 neurons) ---
# Shape of W1: (3 neurons × 2 inputs)
W1 = np.array([[0.5, -0.3],   # weights for neuron 1
               [0.8,  0.4],   # weights for neuron 2
               [-0.2, 0.9]])  # weights for neuron 3
b1 = np.array([0.1, -0.1, 0.2])   # one bias per neuron

# --- Layer 2 weights and bias (3 inputs → 1 neuron) ---
W2 = np.array([[0.6, 0.3, -0.5]])
b2 = np.array([0.1])

print("=" * 50)
print("Forward Pass")
print("=" * 50)

# --- Layer 1: compute weighted sum, then apply ReLU ---
z1 = W1 @ X + b1          # matrix multiplication: (3×2) @ (2,) = (3,)
a1 = relu(z1)              # apply ReLU activation to each element

print(f"\nInput X:              {X}")
print(f"\nLayer 1 — weighted sums (z1): {np.round(z1, 4)}")
print(f"Layer 1 — after ReLU  (a1): {np.round(a1, 4)}")

# --- Layer 2: compute weighted sum, then apply Sigmoid ---
# Sigmoid is used in the output layer here to produce a value between 0 and 1
z2 = W2 @ a1 + b2
a2 = sigmoid(z2)

print(f"\nLayer 2 — weighted sum (z2): {np.round(z2, 4)}")
print(f"Layer 2 — after Sigmoid (a2): {np.round(a2, 4)}")

print(f"\nNetwork prediction: {a2[0]:.4f}")
print("(Interpret as: predicted probability or scaled score)")
print("\nForward pass complete. No weights changed — this was pure calculation.")
```

## Phase 2: The Loss Function

After the forward pass, we have a prediction. We also have the correct answer (the label from our training data). The **loss function** measures the gap between them — a single number that tells us how wrong the network was.

The goal of training is to make this number as small as possible.

| Loss Function | Used for | Intuition |
|--------------|----------|----------|
| **Mean Squared Error (MSE)** | Regression | Average of squared differences between prediction and truth |
| **Binary Cross-Entropy** | Binary classification (yes/no) | Penalises confident wrong predictions most heavily |
| **Cross-Entropy** | Multi-class classification | Generalises binary cross-entropy to many classes |

```python
# ─────────────────────────────────────────────────────────────────────────────
# Computing the loss
#
# We have a prediction from the forward pass.
# Now we compare it to the true label and compute the loss.
# ─────────────────────────────────────────────────────────────────────────────

# True label: did this student pass? (1 = yes, 0 = no)
true_label = 1.0   # this student actually passed

# Our prediction from the forward pass
prediction = a2[0]

# Binary Cross-Entropy loss
# BCE penalises the model more when it is confident and wrong
# Formula: -[y * log(p) + (1-y) * log(1-p)]
epsilon = 1e-8  # small number to prevent log(0)

def binary_cross_entropy(y_true, y_pred):
    """Full Binary Cross-Entropy loss for a single prediction.
    Args:
        y_true: true label (0 or 1)
        y_pred: predicted probability (between 0 and 1)
    Returns:
        scalar loss value — lower is better
    """
    return -(y_true * np.log(y_pred + epsilon) +
             (1 - y_true) * np.log(1 - y_pred + epsilon))

bce_loss = binary_cross_entropy(true_label, prediction)

print(f"True label:  {true_label}  (student passed)")
print(f"Prediction:  {prediction:.4f}  (network's estimated probability of passing)")
print(f"\nBinary Cross-Entropy Loss: {bce_loss:.4f}")
print()

# Illustrate how loss changes with prediction quality
# We call the same bce_loss function for every prediction value
print("Loss intuition — how BCE responds to different predictions when true label = 1:")
print("-" * 55)
for pred in [0.1, 0.3, 0.5, 0.7, 0.9, 0.99]:
    loss = binary_cross_entropy(y_true=1.0, y_pred=pred)
    bar  = '█' * int(loss * 10)
    print(f"  Prediction: {pred:.2f}  →  Loss: {loss:.3f}  {bar}")

print("\nThe further the prediction is from the truth, the higher the loss.")
print("Backpropagation uses this loss signal to adjust the weights.")
```

## Phase 3: Backpropagation

Backpropagation is the algorithm that takes the loss and works out how much each weight in the network contributed to it — and in which direction to adjust each weight to reduce the loss.

The mathematics involves the **chain rule** of calculus. You do not need to derive it. What matters is the intuition:

> Each weight receives a **gradient** — a number that says: *"if you increase this weight slightly, does the loss go up or down, and by how much?"*

The weight is then adjusted in the direction that reduces the loss. The size of the adjustment is controlled by the **learning rate** — a small number (typically around 0.001) that prevents the network from overreacting to any single batch of data.

This process — forward pass, compute loss, backpropagate, update weights — is called one **training step**. It is repeated thousands or millions of times until the loss stops decreasing meaningfully.

```python
# ─────────────────────────────────────────────────────────────────────────────
# Figure 1.4 — The Training Loop
# Visualises the cycle: Forward Pass → Loss → Backprop → Update Weights → repeat
# ─────────────────────────────────────────────────────────────────────────────

fig, ax = plt.subplots(figsize=(12, 5))
ax.set_xlim(0, 12)
ax.set_ylim(0, 5)
ax.axis('off')
fig.patch.set_facecolor('white')
ax.set_facecolor('white')

steps = [
    {'x': 1.2,  'label': '1. Load Batch',      'sub': 'Small group of\ntraining examples',   'color': '#d4e6f1', 'edge': '#2980b9'},
    {'x': 3.6,  'label': '2. Forward Pass',     'sub': 'Data flows through\nnetwork → prediction', 'color': '#d5f5e3', 'edge': '#27ae60'},
    {'x': 6.0,  'label': '3. Compute Loss',     'sub': 'Compare prediction\nto true label',    'color': '#fef9e7', 'edge': '#f39c12'},
    {'x': 8.4,  'label': '4. Backpropagate',    'sub': 'Compute gradient\nfor every weight',  'color': '#fdebd0', 'edge': '#e67e22'},
    {'x': 10.8, 'label': '5. Update Weights',   'sub': 'Adjust weights to\nreduce loss',       'color': '#fadbd8', 'edge': '#e74c3c'},
]

for step in steps:
    rect = mpatches.FancyBboxPatch((step['x'] - 1.0, 1.8), 2.0, 1.6,
                                    boxstyle='round,pad=0.1',
                                    facecolor=step['color'], edgecolor=step['edge'], linewidth=2)
    ax.add_patch(rect)
    ax.text(step['x'], 2.95, step['label'], ha='center', va='center',
            fontsize=9, fontweight='bold', color='#1a1a1a')
    ax.text(step['x'], 2.25, step['sub'],   ha='center', va='center',
            fontsize=8, color='#444444', linespacing=1.4)

# Arrows between steps
for i in range(len(steps) - 1):
    x1 = steps[i]['x']   + 1.0
    x2 = steps[i+1]['x'] - 1.0
    ax.annotate('', xy=(x2, 2.6), xytext=(x1, 2.6),
                arrowprops=dict(arrowstyle='->', color='#555555', lw=2.0))

# Loop-back arrow: step 5 → step 1 (through bottom)
ax.annotate('', xy=(1.2, 1.8), xytext=(10.8, 1.8),
            arrowprops=dict(arrowstyle='->', color='#2E5FA3', lw=2.0,
                            connectionstyle='arc3,rad=0.3'))
ax.text(6.0, 0.7, 'Repeat for every batch, every epoch', ha='center',
        fontsize=9, color='#2E5FA3', fontstyle='italic')

ax.set_title('Figure 1.4 — The Training Loop\nOne cycle of forward pass → loss → backprop → weight update',
             fontsize=12, fontweight='bold', color='#1F3864', pad=10)
plt.tight_layout()
plt.show()

print("Key vocabulary:")
print("  Epoch     = one complete pass through the entire training dataset")
print("  Batch     = a small subset of training examples processed together")
print("  Iteration = one forward + backward pass on one batch")
print("  Gradient  = the direction and magnitude to adjust each weight")
```

```python
# ─────────────────────────────────────────────────────────────────────────────
# Visualising learning: how loss decreases over training epochs
#
# This simulates what a real training loss curve looks like.
# You will see curves like this in every chapter that trains a model.
# ─────────────────────────────────────────────────────────────────────────────

np.random.seed(42)

epochs = 50

# Simulate a typical training and validation loss curve
# Training loss decreases smoothly
train_loss = 1.0 * np.exp(-0.08 * np.arange(epochs)) + np.random.normal(0, 0.015, epochs)
train_loss = np.clip(train_loss, 0.05, None)

# Validation loss decreases but levels off and may tick up slightly (overfitting)
val_loss = 1.05 * np.exp(-0.07 * np.arange(epochs)) + np.random.normal(0, 0.025, epochs)
val_loss[35:] += np.linspace(0, 0.05, 15)  # slight uptick at the end
val_loss = np.clip(val_loss, 0.08, None)

fig, ax = plt.subplots(figsize=(10, 5))
ax.plot(range(1, epochs + 1), train_loss, color='#2980b9', lw=2.5, label='Training loss')
ax.plot(range(1, epochs + 1), val_loss,   color='#e74c3c', lw=2.5, linestyle='--', label='Validation loss')

# Annotate key regions
ax.axvspan(1,  15,  alpha=0.08, color='green',  label='Learning rapidly')
ax.axvspan(15, 35,  alpha=0.08, color='blue',   label='Improving steadily')
ax.axvspan(35, 50,  alpha=0.08, color='red',    label='Risk of overfitting')

ax.text(8,  0.92, 'Learning\nrapidly',   ha='center', fontsize=9, color='#1e8449')
ax.text(25, 0.55, 'Improving\nsteadily', ha='center', fontsize=9, color='#1a5276')
ax.text(43, 0.35, 'Risk of\noverfitting', ha='center', fontsize=9, color='#922b21')

ax.set_xlabel('Epoch', fontsize=11)
ax.set_ylabel('Loss', fontsize=11)
ax.set_title('Figure 1.5 — Typical Training and Validation Loss Curves\n'
             'Loss decreases as the network learns from data',
             fontsize=12, fontweight='bold', color='#1F3864')
ax.legend(fontsize=10)
plt.tight_layout()
plt.show()

print("Reading this plot:")
print("  Training loss (blue):    how well the model fits the data it was trained on")
print("  Validation loss (red):   how well the model generalises to unseen data")
print("  When val loss rises while train loss falls: the model is overfitting")
print("  Overfitting and how to prevent it is covered in depth in Chapter 3.")
```
